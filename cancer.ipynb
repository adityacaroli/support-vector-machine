```python
# outliers
```

```python
fig = plt.figure(figsize = (20,20))
sns.set()
sns.boxplot(data=df[['mean radius', 'mean perimeter', 'mean area', 'mean compactness',
       'mean concavity', 'mean concave points', 'radius error',
       'perimeter error', 'area error', 'worst radius', 'worst perimeter',
       'worst area', 'worst compactness', 'worst concavity',
       'worst concave points']], palette='rainbow', orient='h');
```

```python
def outlier(df, col):
    q3 = df[col].describe()['75%']
    q1 = df[col].describe()['25%']
    iqr = q3-q1
    maxx = q3 + (1.5*iqr)
    minn = q1 - (1.5*iqr)
    df[col] = df[col].apply(lambda x: q3 if x>=maxx else q1 if x<=minn else x)
```

```python
cols = ['mean radius', 'mean perimeter', 'mean area', 'mean compactness',
       'mean concavity', 'mean concave points', 'radius error',
       'perimeter error', 'area error', 'worst radius', 'worst perimeter',
       'worst area', 'worst compactness', 'worst concavity',
       'worst concave points']

for col in cols:
    outlier(df, col) 
```

```python
# train test split
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=41)
```

```python
# scaling/ scale the data
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

sc = StandardScaler() # makes the data normally distributed with unit variance and zero mean
mc = MinMaxScaler()   # scales value from range 0 to 1
```

```python
X_train = mc.fit_transform(X_train)
X_test = mc.transform(X_test)
```

```python
# support vector classifier
```

```python
from sklearn.svm import SVC

model = SVC(gamma='scale')
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print('Accuracy Score: ', round(accuracy_score(y_test, pred)*100,2)) 
print('\n')
print('Classification Report: ', '\n', '\n', classification_report(y_test, pred))
print('\n')
print('Confusion Matrix: ', '\n', confusion_matrix(y_test, pred))
```

```python
'''
True Positive (TP) => actually positive & predicted positive too
                   => actually having cancer & predicted cancerous too
True Negative (TN) => actually negative & predicted negative too
                   => actually not having cancer & predicted not cancerous too
False Positive (FP) => falsely positive info
                    => actually didnt have cancer but predicted cancerous
                    => also known as type 1 error
False Negative (FN) => falsely negative info
                    => actually having cancer but predicted not cancerous
                    => also known as type 2 error
                    => eg. to remember: pregnant lady (actually pregnant but predicted not pregnant)
                    
Accuracy => (TP + TN)/ (TOTAL)
         => (TP + TN)/ (TP + TN + FP + FN)
         => basically means overall how often it is correct
         
         
Recall => more important than accuracy
       => TP/ (TP + FN) 
       => these are basically total cases having cancer (actually positive)

Precision => TP/ (TP + FP)
          => basically tells us of people predicted cancerous, how many actually had cancer
                                                                              (predicted postive)
F1-Score => (2 * Recall * Precision)/ (Precision + Recall)

order of checking => f1 score > recall > precision > accuracy


Confusion Matrix:
                                Actual Positive     Actual Negative
    Predicted Positive                TP                   FP
    Predicted Negative                FN                   TN

'''
```
